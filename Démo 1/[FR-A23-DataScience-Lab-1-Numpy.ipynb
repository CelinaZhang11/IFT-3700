## NUMPY

NumPy (**Num**erical **Py**thon) est une librairie Python utilisé pour travailler avec des objects multi-dimensionnels comme des tableaux et des matrics. On l'utilise pour d'autre librairies commen Pandas, Scipy, Matplotlib, scikit-learn.

**Pourquoi NumPy?**
* L'exécution est extrêment rapide grâce à l'utilisation de code C précompilé Execution is extremely fast by using a precompiled C code
* On peut éviter les boucles en utilisation la vectorization
* Facile à l'utiliser

La documentation des fonctions NumPy se trouve [ici.](https://docs.scipy.org/doc/numpy/reference/)

### [5] Diffuser

Diffuser décrit comment NumPy traite les tableaux avec dimensions différentes durant des opérations arithmetiques. Tant que certaines contraintes sont respectées, le tableau plus petit est "diffusé" le long du plus grand tableau pour qu'ils ont une taille compatible.

La plupart des opérations Numpy sont executées sur une paire de tableaux élément par élément. Pour faire cela, NumPy compare les formes des 2 tableaux, commençant avec la dimension la plus à droite.
**2 dimensions sont compatibles quand:**

- ils sont égales, ou
- une des dimensions est 1

Transformation géométrique simple

Note : Ignorez la partie sur la création de graphique, on parlera de ça plus tard dans le cours




````python
from matplotlib import pyplot as plt
````

````python
coordinates = np.array([[0,3], [5,3], [5,0], [0,0]])
plt.scatter(*zip(*coordinates))
````

````python
scale = np.array([[0.7,0],[0,0.5]])
coordinates_scaled = np.matmul(scale, np.transpose(coordinates))
coordinates_scaled = np.transpose(coordinates_scaled)
plt.scatter(*zip(*coordinates_scaled))
````
